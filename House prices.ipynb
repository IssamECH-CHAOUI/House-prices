```python
"x est une variable qualitative et y est une variable quantitative"

def rapport_corr(x,y):
    somme = 0
    
    for facteur in x.value_counts().index:
        ind = x[x == facteur].dropna(axis=0).index
        classe = y.iloc[ind,]
        somme += len(classe)*np.sum((np.mean(classe)-np.mean(y))**2)
        
    y_ecart = np.sum((y-np.mean(y))**2)    
    rc = somme/y_ecart
    
    return rc
```
